```python
%cd ~/change
import pickle
from fugashi import Tagger
import re
tagger = Tagger('-Owakati')
```

```python
f = open('data/mopen.txt')
lines = f.readlines()
splited = []
for l in lines:
    l = l.replace("。 ",'@')
    splited.append(l.split("@"))
train_y = [0] * len(splited)
train_x = [[0] * 4 for i in range(len(splited))]
```

```python
for i in range(len(splited)):
    train_x[i][0] = splited[i][6]
    train_x[i][1] = splited[i][3][:splited[i][3].rfind(splited[i][4])]
    train_x[i][2] = splited[i][5]
    train_x[i][3] = splited[i][4]
    train_y[i] = splited[i][0]
```

```python
#use se_用全単　and fugashi
for i in range(len(splited)):
    if('s4_' in lines[i]):
        m = re.search(r's4_用全単：(.+?)\s',lines[i])
        train_x[i][2] = m.group(1)
    else:
        train_x[i][2] = tagger(train_x[i][2])[0].feature.lemma
        print(i)
```

```python
for i in range(len(train_x)):
    index = train_x[i][0].find(' ')
    if(index != -1):
        train_x[i][0] = train_x[i][0][:index]
```

```python
for i in range(len(train_x)):
    print(train_x[i],end='')
    print('\t'  + train_y[i])
```

```python
pat = re.compile('[!"#$%&\'\\\\()*+,-./:;<=>?@[\\]^_`{|}~「」〔〕“”〈〉『』【】＆＊・（）＄＃＠。、？！｀＋￥％]')
for i in range(len(splited)):
    train_x[i][0] = pat.sub('',train_x[i][0])
    train_x[i][1] = pat.sub('',train_x[i][1])
    train_x[i][2] = pat.sub('',train_x[i][2])
    train_y[i] = pat.sub('',train_y[i])
```

```python
with open('data/mopen.pickle',mode='wb') as f:
    pickle.dump(train_x,f)
    pickle.dump(train_y,f)
```
